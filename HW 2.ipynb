```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
MPG = pd.read_csv('/Users/21car/Downloads/MPG.csv')
print (MPG.head())
```

```python
avg_mpg_by_year = MPG.groupby('model year')['mpg'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='model year', y='mpg', data=avg_mpg_by_year)
plt.title('Average MPG by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
```

```python
print('It seems like the graph trends to higher average MPG as the years prorgress, the trend does not consistently go up every year though, and there is some variation')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(MPG['horsepower'], MPG['mpg'])
plt.title('Scatter Plot of MPG vs. Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```

```python
print('There seems to be a downward coorelation between MPG and horsepower')
```

```python
plt.figure(figsize=(10, 6))
plt.bar(MPG['origin'], MPG['mpg'])
plt.title('Bar Chart of MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('MPG')
plt.show()
```

```python
print('Based off this graph, it looks like as orgin goes up, so does MPG')
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='origin', hue='car manufacturer', data=MPG)
plt.title('Bar Chart of Origin by Car Manufacturer')
plt.xlabel('Origin')
plt.ylabel('Count')
plt.legend(title='Car Manufacturer', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
origin_percentage = MPG['origin'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 5))
origin_percentage.sort_index().plot(kind='bar', color='skyblue')
plt.title('Percentage of Cars in Each Origin')
plt.xlabel('Origin')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
print(':The dataset appears to majorly consist of the first orgin, with almost 60% of the dataset being form there')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(MPG['mpg'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
print('The four highest frequencies in this graph are all back to back, as 13 through 20 seems to be the most frequent occuring MPG. The majority of occuring frequencies are 12 MPG through 35 MPG.')
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='cylinders', y='mpg', data=MPG, palette='Set2')
plt.title('Boxplot of MPG by Cylinder Numbers')
plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
print('4 Cylinders appears to be the most effective, with 5 being close behind. 3, 6, and 8 cylinders all are much worse with MPG')
```

```python
corrMatrix = MPG.corr()
sns.heatmap(corrMatrix)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='weight', y='mpg', hue='origin', data=MPG, palette='Set1')
plt.title('Scatter Plot of MPG vs. Weight with Origin Colors')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
print('The graph shows that MPG goes down as Weight of the car increases')
```

```python
selected_columns = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']

sns.pairplot(MPG[selected_columns])
plt.suptitle('Pairplot of MPG, Displacement, Horsepower, Weight, and Acceleration', y=1.02)
plt.show()
```

```python
print('MPG appears to go down as weight horespower displacement go up, and MPG goes up as accelaration gets better. Displacement goes up when weight and horespower increase, and goes down when acc goes down. As weight increases, so does horespwer, and acc decreases.')
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='horsepower', y='mpg', data=MPG)
plt.title('Scatter Plot of MPG vs. Horsepower (Without Log Transformation)')
plt.xlabel('Horsepower')
plt.ylabel('MPG')

plt.subplot(1, 2, 2)
sns.scatterplot(x='horsepower', y='mpg', data=MPG)
plt.xscale('log')  # Apply log transformation to x-axis
plt.title('Scatter Plot of MPG vs. Horsepower (With Log Transformation)')
plt.xlabel('Log(Horsepower)')
plt.ylabel('MPG')

plt.tight_layout()
plt.show()
```

```python
print('because horsepower usually appears as a non log transformation, I think the graph without the log transformation makes more sense to use.')
```
